### R-3.1 Graph the functions 8n, 4nlogn, 2n2, n3, and 2n using a logarithmic scale for the x- and y-axes; that is, if the function value f(n) is y, plot this as a point with x-coordinate at logn and y-coordinate at logy.

```python
import numpy as np 
import matplotlib.pyplot as plt
import math
```

```python
x_inputs = np.linspace(0, 5, 15)
plt.plot(x_inputs)
plt.plot(x_inputs, 8*x_inputs, x_inputs, 'r-', x_inputs, 4*x_inputs*np.log2(x_inputs), 'bs', x_inputs, np.power(x_inputs, 3), 'g^', x_inputs, np.power(2, x_inputs), 'g+')
plt.ylabel('some numbers')
plt.show()
```

### R-3.2 The number of operations executed by algorithms A and B is 8nlog n and 2n2, respectively. Determine n0 such that A is better than B for n ≥ n0.

```python
x_inputs = np.linspace(0, 30, 10)

plt.plot(x_inputs, 8*x_inputs*np.log2(x_inputs),'r-', x_inputs, 2*np.power(x_inputs, 2),'b-')
plt.xlabel('Input Size')
plt.ylabel('Time')
plt.show()
"""The answer is approximately 16 """ 
```

### R-3.3 The number of operations executed by algorithms A and B is 40n2 and 2n3, respectively. Determine n0 such that A is better than B for n ≥ n0.


```python
x_inputs = np.linspace(0, 30, 10)

plt.plot(x_inputs, 40* np.power(x_inputs, 2),'r-', x_inputs, 2*np.power(x_inputs, 3),'b-')
plt.xlabel('Input Size')
plt.ylabel('Time')
plt.show()
"""The answer is approximately 20 """ 
```
